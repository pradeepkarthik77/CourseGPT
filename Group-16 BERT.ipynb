```python
ans=[]
for i in range(df.shape[0]):
    val=[]
    val.append(df["Course Name"][i])
    val.append(df["Course Description"][i])
    val.append(df["Skills"][i])
    ans.append(" ".join(val))
ans    
```

```python
df["Data"]=ans
df
```

```python
nlp=spacy.load("en_core_web_sm")

def preprocess(text):
    doc=nlp(text)
    filtered_token=[]
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_token.append(token.lemma_)
    return " ".join(filtered_token)

print(preprocess(df["Data"][0]))
    
```

```python
def doeverything(val):
    df1=df[val]
    df1=df1.apply(lambda x: str(x))
    prep=[]
    for i in range(len(df1)):
        prep.append(preprocess(df1[i]))
    return prep
```

```python
preprocessed=doeverything("Data")
```

```python
bert_preprocess = hub.KerasLayer("bert_en_uncased_preprocess_3")
```

```python
bert_encoder=hub.KerasLayer("bert_en_uncased_L-12_H-768_A-12_4")
```

```python
def get_sentence_embeding(sentences):
    preprocessed_text = bert_preprocess(sentences)
    return bert_encoder(preprocessed_text)['pooled_output']
```

```python
user_inp="I am a beginner looking for a Python programming course that covers essential programming elements, basic Python programs, expressions, variables, functions, logic, conditionals, Python modules, and practical tasks. I want to learn how to think like a computer programmer and write Python programs. The course should use Python 3 and preferably provide a beginner-friendly development environment like CodeSkulptor that runs in a web browser. I'm also interested in progressing to more advanced desktop development environments in later courses."
```

```python
def check(user_inp):
    user_inp=preprocess(user_inp)
    sent_embd=get_sentence_embeding([user_inp])
    return sent_embd
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sent_embd=check(user_inp)
simm=[]
for i in range(len(val)):
    simm.append(cosine_similarity([val[i]],sent_embd))
#print(simm[:10])
final=simm[:]
final.sort(reverse=True)
ans=[]
for i in range(40):
    check_val=final[i]
    #print(check_val)
    for j in range(len(val)):
        if(simm[j]==check_val):
            #print(simm[j])
            ans.append(df["Course Name"][j])

print(ans)
    
```
